```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel (r'C:\Users\Rafi Sunggoro\VizBizz/Belajar VizBizz/3.PR/PR04/JII-ANLZ-2021.xlsx')
data.head(30)
```

```python
data['Liner'] = pd.cut(data['Market Cap'],
                         bins = [0,500000000000,10000000000000,999999999999999999],
                         labels = ['Third Liner', 'Second Liner', 'First Liner / Blue Chip'])
data['Value'] = pd.cut(data['PBVxPER'],
                         bins = [-99999,22.5,999999],
                         labels = ['Undervalue', 'Overvalue'])
data['Yield Ideal'] = pd.cut(data['Yield'],
                         bins = [-99999,0.07,999999],
                         labels = ['Not ideal', 'Ideal'])
data['Profit'] = pd.cut(data['NPM'],
                         bins = [-99999,0.1,999999],
                         labels = ['Non Profit', 'Profit'])
data['Management'] = pd.cut(data['OPM'],
                         bins = [-99999,0.2,999999],
                         labels = ['Uneffecient Management', 'Effecient Management'])
data['Return On Equity'] = pd.cut(data['ROE'],
                         bins = [-99999,0.15,999999],
                         labels = ['Not Ideal', 'Ideal'])
data['Debt to Equity'] = pd.cut(data['DER'],
                         bins = [-99999,1,999999],
                         labels = ['Ideal', 'Not Ideal'])

df = pd.DataFrame(data)
df.head()
```

```python
df1= pd.DataFrame(data, columns =['Kode', 'PBVxPER', 'Value'])
value = df1.loc[df1['Value'] == 'Undervalue']
topvalue = value.nsmallest(3, ['PBVxPER'])
x = topvalue['Kode']
y = topvalue['PBVxPER']
plt.figure(figsize=(6,5))
plt.grid(color='white', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.bar(x=np.arange(len(x)),height=y)
plt.xticks(np.arange(len(x)), x, rotation=0, color='white')
plt.yticks(color='white')
plt.xlabel('Ticker', color='white')
plt.ylabel('Value', color='white')
plt.title('Top 3 Undervalued Companies - JII Stocks', color='white')
plt.savefig('Undervalued.png', transparent=True)

plt.show()
```

```python
df2= pd.DataFrame(data, columns =['Kode', 'Yield', 'Yield Ideal'])
dividend = df2.loc[df2['Yield Ideal'] == 'Ideal']
topdividend = dividend.nlargest(3, ['Yield'])
x = topdividend['Kode']
y = topdividend['Yield']
plt.figure(figsize=(6,5))
plt.grid(color='white', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.bar(x=np.arange(len(x)),height=y)
plt.xticks(np.arange(len(x)), x, rotation=0, color='white')
plt.yticks(color='white')
plt.xlabel('Ticker', color='white')
plt.ylabel('Dividend Yield', color='white')
plt.title('Top 3 Companies with an Ideal Dividend Yield - JII Stocks', color='white')
plt.savefig('Dividend Yield.png', transparent=True)

plt.show()
```

```python
df3= pd.DataFrame(data, columns =['Kode', 'NPM', 'Profit'])
profit = df3.loc[df3['Profit'] == 'Profit']
topprofit = profit.nlargest(3, ['NPM'])
x = topprofit['Kode']
y = topprofit['NPM']
plt.figure(figsize=(6,5))
plt.grid(color='white', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.bar(x=np.arange(len(x)),height=y)
plt.xticks(np.arange(len(x)), x, rotation=0, color='white')
plt.yticks(color='white')
plt.xlabel('Ticker', color='white')
plt.ylabel('Net Profit Margin', color='white')
plt.title('Top 3 Profitable Companies - JII Stocks', color='white')
plt.savefig('Profitable Company.png', transparent=True)

plt.show()
```
